```python
import nltk
from nltk import word_tokenize
from newspaper import Article
import matplotlib.pyplot as plt
```

```python
url = 'https://www.educationdive.com/news/how-one-company-works-with-colleges-and-tech-firms-to-bridge-the-skills-gap/558332/?fbclid=IwAR2VI8_qugdxVCV-CgtSeAf0ggILJEECRXg8F1wQzvVRgYTpSd6PMpxYl8Y'
dive_article = Article(url, language='en')
```

```python
dive_article.download()
```

```python
dive_article.parse()
```

```python
dive_article.nlp()
```

```python
print("Article Text: ")
article_body = dive_article.text
print(article_body)
```

```python
tokens = [t for t in article_body.split()]
print(tokens)
```

```python
word_tokens = word_tokenize(article_body)
print(word_tokens)
```

```python
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
freq.plot(20, cumulative=False)
```
